```python
# Exercise 1: Prime Numbers
def is_prime(n):
    """Check if a number is prime."""
    if n <= 1:
        return False  # Numbers less than or equal to 1 are not prime
    if n == 2:
        return True  # 2 is the only even prime number
    if n % 2 == 0:
        return False  # Other even numbers are not prime
    for i in range(3, int(n**0.5) + 1, 2):  # Check divisibility from 3 to sqrt(n), skipping even numbers
        if n % i == 0:
            return False  # If divisible by any number, it is not prime
    return True  # If no divisors found, it is prime

# Test the function with an example
number = int(input("Enter a number to check if it is prime: "))
if is_prime(number):
    print(f"{number} is a prime number.")
else:
    print(f"{number} is not a prime number.")
```

```python
# Exercise 2: Product of Random Numbers
import random

def check_product():
    # Generate two random numbers between 1 and 100
    num1 = random.randint(1, 100)
    num2 = random.randint(1, 100)
    
    # Calculate the product of the two numbers
    product = num1 * num2
    
    # Ask the user to enter the product
    user_input = int(input(f"What is the product of {num1} and {num2}? "))
    
    # Check if the user's answer is correct
    if user_input == product:
        print("Correct! Well done.")
    else:
        print(f"Incorrect. The correct answer is {product}.")

# Test the function
check_product()
```

```python
# Exercise 3: Squares of Even/Odd Numbers
def print_squares_of_even_numbers(start, end):
    """
    Print the squares of all even numbers within the given range.
    Chosen to work with even numbers.
    """
    for num in range(start, end + 1):
        if num % 2 == 0:  # Check if the number is even
            print(f"The square of {num} is {num**2}")

# Define the range
start_range = 100
end_range = 200

# Call the function to print squares of even numbers
print_squares_of_even_numbers(start_range, end_range)
```

```python
# Exercise 4: Word Counter
def word_counter_exact(text):
    # Split the text into words
    words = text.split()
    
    # Create a dictionary to store word counts
    word_count = {}
    
    # Count the occurrences of each word
    for word in words:
        if word in word_count:
            word_count[word] += 1  # If word already exists, increment its count
        else:
            word_count[word] = 1  # If word does not exist, initialize its count to 1
            
    return word_count

# Example input text
input_text = "This is a sample text. This text will be used to demonstrate the word counter."

# Get the word count
result = word_counter_exact(input_text)

# Print the word counts
for word, count in result.items():
    print(f"'{word}': {count}")
```

```python
# Exercise 5: Check for Palindrome
def is_palindrome(s):
    """
    Check if the given string is a palindrome.
    
    A palindrome reads the same forward and backward, ignoring spaces, punctuation, and capitalization.
    """
    # Normalize the string: remove spaces, punctuation, and convert to lowercase
    import string
    normalized_s = ''.join(char.lower() for char in s if char not in string.punctuation and char != ' ')
    
    # Check if the normalized string is equal to its reverse
    return normalized_s == normalized_s[::-1]

# Test cases
test_strings = ["racecar", "A man, a plan, a canal, Panama!", "No lemon, no melon", "hello"]

# Print results
for test in test_strings:
    print(f"Input: '{test}'\nIs palindrome: {is_palindrome(test)}\n")
```
